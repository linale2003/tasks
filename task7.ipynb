```python
import pandas as pd 
import numpy as np 
import pylab
```

```python
cols=['name','gender','birth']
years= range(1880,2011)
pieces=[]
for i in years:
    df=pd.read_table('C:\\Users\\79247\\babynames\\yob%d.txt'%i, sep = ',', engine = 'python' , names = cols)
    df["year"]=i
    pieces.append(df)
    data = pd.concat(pieces,ignore_index=True)
fasd=data.pivot_table('birth',index='gender',aggfunc='sum')
fasd
```

```python
import matplotlib.pyplot as plt
op=data.pivot_table('birth',index='year',columns='gender',aggfunc='sum')
op
a=op['F']
b=op['M']
c=years
plt.plot(c,a,label="Female",color='black')
plt.plot(c,b,label="Male",color='red')
plt.legend()
plt.show()
```

```python
data['proportion']=data['birth']/sum(data['birth'])
data
```

```python
name= data.pivot_table('birth', index='year', columns='name', aggfunc='sum')
pylab.subplot (2, 2, 1)
pylab.plot(c, name['Johnny'], color = 'black')
pylab.title('Johnny')
pylab.subplot (2, 2, 2)
pylab.plot(c, name['Natalie'], color = 'red')
pylab.title('Natalie')
pylab.subplot (2, 2, 3)
pylab.plot(c, name['Bob'], color = 'green')
pylab.title('Bob')
pylab.subplot (2, 2, 4)
pylab.plot(c, name['Nikita'], color = 'blue')
pylab.title('Nikita')
plt.subplots_adjust(wspace = 0.5, hspace = 0.5)
```
