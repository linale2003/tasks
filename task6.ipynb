```python
import pandas as pd
from pandas import Series, DataFrame
```

```python
movies = pd.read_table('C:\Bar_3\Study_programming\pandas\movies.dat', sep='::', names=['MovieID', 'Title', 'Tag'], engine='python')
movies
```

```python
users = pd.read_table('C:\\Bar_3\\Study_programming\\pandas\\users.dat', sep='::', names = ['UserID', 'Gender', 'Age', 'Occepation', 'ZipCode'], engine='python')
users
```

```python
ratings = pd.read_table('C:\\Bar_3\\Study_programming\\pandas\\ratings.dat', sep='::', names=['UserID', 'MovieID', 'Rating', 'Timestap'], engine='python')
ratings
```

```python
data = pd.merge(ratings, users)
data
```

```python
data = pd.merge(data, movies)
data
```

```python
mean_ratings_gender = data.pivot_table('Rating', index='Title', columns='Gender', aggfunc='mean')
mean_ratings_gender
```

```python
mean_ratings_gender_sorted = mean_ratings_gender.sort_values(by='F', ascending=False)
mean_ratings_gender_sorted
```

```python
mean_ratings_age = data.pivot_table('Rating', index='Title', columns='Age', aggfunc='mean')
mean_ratings_age
```

```python
mean_ratings_age_sorted = mean_ratings_age.sort_values(by=56, ascending=False)
mean_ratings_age_sorted
```
